```python
import tkinter as tk
from tkinter import ttk, messagebox
import openpyxl
from tkinter import filedialog
from datetime import datetime
import os

class AttendanceApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Attendance App")

        self.rows = 5
        self.cols = 10
        self.students = [f"Roll No {i}" for i in range(1, self.rows * self.cols + 1)]
        self.attendance_status = {student: tk.BooleanVar() for student in self.students}

        self.create_widgets()

    def create_widgets(self):
        for row in range(self.rows):
            for col in range(self.cols):
                index = row * self.cols + col
                student = self.students[index]
                check_button = ttk.Checkbutton(self.root, text=student, variable=self.attendance_status[student])
                check_button.grid(row=row, column=col, padx=5, pady=5, sticky=tk.W)

        submit_button = ttk.Button(self.root, text="Submit", command=self.submit_attendance, style="Accent.TButton")
        submit_button.grid(row=self.rows, columnspan=self.cols, pady=10)

    def submit_attendance(self):
        attendance_matrix = [[1 if self.attendance_status[student].get() else 0 for student in self.students[i:i+self.cols]] for i in range(0, len(self.students), self.cols)]

        present_count = sum(sum(row) for row in attendance_matrix)
        absent_count = len(self.students) - present_count

        attendance_str = "\n".join([" ".join(map(str, row)) for row in attendance_matrix])
        message = f"Attendance Matrix:\n{attendance_str}\n\nPresent Students: {present_count}\nAbsent Students: {absent_count}"
        messagebox.showinfo("Attendance", message)

        self.save_to_excel(attendance_matrix)

    def save_to_excel(self, attendance_matrix):
        today = datetime.now()
        month = today.strftime("%B")
        year = today.year

        if not os.path.exists("attendance_records"):
            os.makedirs("attendance_records")

        file_path = f"attendance_records/{month}_{year}.xlsx"
        workbook = openpyxl.Workbook()

        if os.path.exists(file_path):
            workbook = openpyxl.load_workbook(file_path)

        if month not in workbook.sheetnames:
            sheet = workbook.create_sheet(month)
            sheet.append(["Date"] + self.students)  # Add header row with student names
        else:
            sheet = workbook[month]

        date = today.strftime("%Y-%m-%d")
        sheet.append([date] + [1 if self.attendance_status[student].get() else 0 for student in self.students])

        for row in sheet.iter_rows(min_row=2, min_col=2, values_only=True):
            if sum(row[1:]) == 0:  # If all attendance values are 0, delete the row
                sheet.delete_rows(row[0] + 1)

        workbook.save(file_path)
        messagebox.showinfo("Excel Saved", f"Attendance data saved to {file_path}")

if __name__ == "__main__":
    root = tk.Tk()
    app = AttendanceApp(root)
    style = ttk.Style()
    style.configure("Accent.TButton", background="green", foreground="white")
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import openpyxl
from datetime import datetime

class AttendanceApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Attendance App")

        self.rows = 5
        self.cols = 10
        self.students = [f"Roll No {i}" for i in range(1, self.rows * self.cols + 1)]
        self.attendance_status = {student: tk.BooleanVar() for student in self.students}

        self.create_widgets()

    def create_widgets(self):
        # Create a label and an entry field for entering the date
        date_label = ttk.Label(self.root, text="Date:")
        date_label.grid(row=self.rows, column=0, padx=5, pady=5, sticky=tk.E)
        self.date_entry = ttk.Entry(self.root)
        self.date_entry.grid(row=self.rows, column=1, padx=5, pady=5, sticky=tk.W)

        for row in range(self.rows):
            for col in range(self.cols):
                index = row * self.cols + col
                student = self.students[index]
                check_button = ttk.Checkbutton(self.root, text=student, variable=self.attendance_status[student])
                check_button.grid(row=row, column=col, padx=5, pady=5, sticky=tk.W)

        submit_button = ttk.Button(self.root, text="Submit", command=self.submit_attendance, style="Accent.TButton")
        submit_button.grid(row=self.rows + 1, columnspan=self.cols, pady=10)

    def submit_attendance(self):
        date = self.date_entry.get()  # Get the date from the entry field
        attendance_matrix = [[date] + [1 if self.attendance_status[student].get() else 0 for student in self.students[i:i + self.cols]] for i in range(0, len(self.students), self.cols)]

        present_count = sum(sum(row[1:]) for row in attendance_matrix)
        absent_count = len(self.students) - present_count

        attendance_str = "\n".join([" ".join(map(str, row)) for row in attendance_matrix])
        message = f"Attendance Matrix:\n{attendance_str}\n\nPresent Students: {present_count}\nAbsent Students: {absent_count}"
        messagebox.showinfo("Attendance", message)

        self.save_to_excel(attendance_matrix)

    def save_to_excel(self, attendance_matrix):
        workbook = openpyxl.Workbook()
        sheet = workbook.active

        # Add headers
        headers = ["Date"] + self.students
        sheet.append(headers)

        for row in attendance_matrix:
            sheet.append(row)

        file_name = "attendance.xlsx"
        workbook.save(file_name)
        messagebox.showinfo("Excel Saved", f"Attendance data saved to {file_name}")

if __name__ == "__main__":
    root = tk.Tk()
    app = AttendanceApp(root)
    style = ttk.Style()
    style.configure("Accent.TButton", background="green", foreground="blue")
    root.mainloop()
```
